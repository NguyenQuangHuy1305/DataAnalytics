```python
import requests, csv
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def get_movies(url):    
    response = requests.get(url)
    soup_imdb = BeautifulSoup(response.text)

    #get movie title
    all_titles = soup_imdb.find_all("td", {"class":"titleColumn"})
    titles = []
    for title in all_titles:
        #cleaning titleColumn
        title = title.get_text().replace('\n', "")
        title = title.strip(" ")

        #since the cleaned titleColumn contain both ranking and title, split it at '.', creating a list
        title = title.split('.')
        #movie title is the 2nd element in the splitted string
        title = title[1]
        #get rid of movie release year
        title = title[0 : title.find('(')]
        titles.append(title)

    # get ranking
    all_titles = soup_imdb.find_all("td", {"class":"titleColumn"})
    rankings = []
    for title in all_titles:
        #cleaning titleColumn
        title = title.get_text().replace('\n', "")
        title = title.strip(" ")

        #since the cleaned titleColumn contain both ranking and title, split it at '.', creating a list
        title = title.split('.')
        #append the 1st element of the splitted list (title) as movie's ranking
        rankings.append(title[0])


    # get release year
        # method1: get year from movie title
        # all_titles = soup_imdb.find_all("td", {"class":"titleColumn"})
        # years = []
        # for title in all_titles:
        #     #cleaning titleColumn
        #     title = title.get_text().replace('\n', "")
        #     title = title.strip(" ")
        #     year = title[title.find('(')+1: title.find(')')]
        #     years.append(year)

        # method2: get year from <span>, class: secondaryInfo
    all_years = soup_imdb.find_all("span", {"class":"secondaryInfo"})
    years = []
    for year in all_years:
        release_year = year.get_text()
        release_year = release_year[release_year.find('(')+1: release_year.find(')')]
        years.append(release_year)

    # get castings
    all_titles = soup_imdb.find_all("td", {"class":"titleColumn"})
    castings = []
    for title in all_titles:
        casting = title.find("a")
        casting = casting['title']
        castings.append(casting)

    # get rating
    all_ratings = soup_imdb.find_all("td", {"class":"ratingColumn imdbRating"})
    ratings = []
    for rating in all_ratings:
        rating = rating.get_text().replace('\n', "")
        ratings.append(rating)

    data = pd.DataFrame()
    data['Title'] = titles
    data['Ranking'] = rankings
    data['Initial release year'] = years
    data['Casting'] = castings
    data['Rating'] = ratings

    # write csv
    data.to_csv('imdb_top_250.csv', index=False)
```

```python
def main():
    url = "https://www.imdb.com/chart/top"
    get_movies(url)
```

```python
main()
```
